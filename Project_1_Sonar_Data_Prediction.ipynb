## Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data Collection and Pre-processing

```python
#loading the dataset to a pandas dataframe
sonar_data = pd.read_csv("/content/Copy of sonar data (1).csv", header=None) #header is set to none as we dont have name for our columns
```

```python
#displaying the first few data from the data set
sonar_data.head()
```

```python
# no of rows and columns in our data set
sonar_data.shape
# intotal we have 61 columns and 208 rows of data
# 60 columns are our features and last one is our target column
```

```python
# some more statistical information from our data
sonar_data.describe()
```

```python
# count of data for rock and mine
#as we dont have a column name hence we passed the column number of our target column
sonar_data[60].value_counts()
#total columns are 61, but as indexing starts from 0 , we have to give index number
# and as we can see from the output is fairly distributed
```

```python
# group the data based on mine and rock
# looking for mean value
sonar_data.groupby(60).mean()
```

```python
# seperating features and target values

X = sonar_data.drop(columns=60, axis = 1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

## Splitting into test and train data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, random_state=1, stratify=Y)

# first we pass our feature x, and label y
# that we said test_size = 0.1 , thet means use 90 percent as training data and 10 percent for test data
# stratify column indicates that the proportion of labels in both training and test data must be same as that of entire dataset
# random state 1 , is just so that we split the data similiarly every time we run the program



```

```python
# lets see the count of data
print(X.shape, X_train.shape, X_test.shape )
print(Y.shape, Y_train.shape, Y_test.shape )


print(X_train)
print(Y_train)




```

## Model Training (Logistic Regression)

```python
# saving an instance of the model in a variable
model = LogisticRegression()

#training the model with training data
# we use the fit function for this purpose

model.fit(X_train, Y_train)










```

## Model Evaluation

```python
# prediction for train and test data

Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

# accuracy on test data of the model
train_data_accuracy = accuracy_score(Y_train_pred, Y_train)
train_data_accuracy = round(train_data_accuracy,2) #rounding off data upto two decimals
print("Accuracy of training data :", train_data_accuracy*100, '%')

# accuracy on test data of the model
test_data_accuracy = accuracy_score(Y_test, Y_test_pred)
test_data_accuracy = round(test_data_accuracy,2)
print("Accuracy of test data :", test_data_accuracy*100, '%')
```
